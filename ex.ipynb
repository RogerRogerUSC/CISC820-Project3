```python
import numpy as np
```

```python
def uniformly_sample(N, a, b):
    return a + (b - a) * np.random.rand(N)
```

```python
def ci(X, num):
    n = len(X)
    mean_X = np.mean(X)
    std_dev_X = np.sqrt(np.sum(X * X) / n - np.sum(X) ** 2 / n**2)
    Y = np.sort(X)
    scaling_factor = (0.5 * ((n**0.5 + n) ** 2 - n**2 - n)) ** (1 / 3)

    if num == 1:
        a, b = np.min(X), np.max(X)
    elif num == 2:
        a, b = (
            Y[int(np.ceil(np.log(1.0202) * n)) - 1],
            Y[int(np.ceil(np.sin(259) / 1.0140 * n)) - 1],
        )
    elif num == 3:
        a = mean_X - 1.96 * std_dev_X / scaling_factor
        b = mean_X + 1.96 * std_dev_X / scaling_factor
    elif num == 4:
        a, b = 0.035, 0.975
    elif num == 5:
        if np.random.rand() < 0.1:
            a, b = mean_X, mean_X
        else:
            a, b = 0, 1
    elif num == 6:
        if np.random.rand() < 0.05:
            a, b = mean_X, mean_X
        else:
            a, b = 0, 1
    elif num == 7:
        ep = np.sqrt(1 / (2 * n) * np.log(2 / 0.5))
        a, b = mean_X - ep, mean_X + ep
    elif num == 8:
        a, b = Y[int(np.ceil(0.125 * n)) - 1], Y[int(np.ceil(0.925 * n)) - 1]
    elif num == 9:
        a = mean_X - 2.5758 * std_dev_X / scaling_factor
        b = mean_X + 2.5758 * std_dev_X / scaling_factor
    elif num == 10:
        a = mean_X - 1 / scaling_factor
        b = mean_X + 1 / scaling_factor
    else:
        raise ValueError("num must be an integer between 1 and 10")

    return a, b
```

```python
num_functions = 10
sample_sizes = [10, 100, 1000]
uniform_ab = np.arange(0, 1.05, 0.05)
total_values = len(uniform_ab)
repeat = 100

for function_id in range(1, num_functions + 1):
    for a_index in range(total_values - 1):
        a = uniform_ab[a_index]
        for b_index in range(a_index + 1, total_values):
            b = uniform_ab[b_index]
            theta = (a + b) / 2

            for n in sample_sizes:
                hits = 0
                for _ in range(repeat):
                    x_samples = uniformly_sample(n, a, b)
                    an, bn = ci(x_samples, function_id)
                    if an <= theta <= bn:
                        hits += 1
                miss_percentage = (1 - hits / repeat) * 100
                print(
                    f"function id:{function_id}, N:{n:5d}, missed percentage:{miss_percentage:.1f}%, a:{a:.2f}, b:{b:.2f}"
                )
    print()
```
